`#` issue1

> convert columns [IMSI, IMEI,  MSISDN/Number] from number type to object/string type

```python
clean_data.IMEI = clean_data.IMEI.astype('object')
clean_data.IMSI = clean_data.IMSI.astype('object')
clean_data['MSISDN/Number'] = clean_data['MSISDN/Number'].astype('object')
```

> Test

```python
test1 = clean_data [['IMSI','IMEI','MSISDN/Number']]
test1.dtypes
```

`#` issue 2 

   a. Outliers needs to be treated

   b. Columns with missing values needs to be filled5

> Treat outliers 

```python
Q1 = clean_data.quantile(0.25)
Q3 = clean_data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

> get columns with outlier values

```python
num_data2 = clean_data.select_dtypes(include='number')
```

```python
outliers =  pd.DataFrame(((((num_data2 < (Q1 - 1.5 * IQR)) |(num_data2 > (Q3 + 1.5 * IQR)))==True)==True)
                         .any()).reset_index()
```

```python
outliers
```

```python
outliers_cols = outliers['index'].tolist()
```

```python
outliers=outliers[outliers[0]==True]
```

```python
outliers_cols
```

We would replace outliers in the dataset with the median of outlier columns

```python
def outlier_replace(df, cols):
    for col in cols:
        df[cols] = np.where(((df[cols] < (Q1 - 1.5 * IQR)) | (df[cols] > (Q3 + 1.5 * IQR))), df[cols].median(), df[cols])
```

```python
outlier_replace(clean_data, outliers_cols)
```

```python
clean_data
```

> Test for outlier values agaain

```python
(((num_data3 < (Q1 - 1.5 * IQR)) |(num_data3 > (Q3 + 1.5 * IQR)))==True).any()
```
