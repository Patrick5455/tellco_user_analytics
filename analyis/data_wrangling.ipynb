> Detecting Outliers in null columns

This would help determine whether to impute null values using medain or mean

we would select all  numeric null columns and calcualate thier z-score to determine 

this would help us understand better hwo to impute the null values

```python
numeric_data = data[null_cols]
```

```python
numeric_data = numeric_data.select_dtypes(include="number")
```

We have to drop incorrect numeric datatype such as IMSI, IMEI.. this would be dealt with in the cleaning phase

```python
numeric_data=numeric_data.drop(columns=['IMSI','IMEI','MSISDN/Number'], axis=1)#,inplace=True)
```

```python
numeric_data
```

```python
numeric_data.dtypes
```

> We would etect outliers in each fo the column by calculating thier Z-Score

```python
cols = list(numeric_data.columns)
outliers = pd.DataFrame()
#now iterate over the remaining columns and create a new zscore column
for col in cols:
    col_zscore = col + '_zscore'
    outliers[col_zscore] = (numeric_data[col] - numeric_data[col].mean())/numeric_data[col].std(ddof=0)
```

```python
outliers
```

#### values with __Z-score >=3 & Z-score <=-3__ are outliers.

 - If the frequency of such values are larger than 40% in any column, we would condier such column as having large number of outliers

- Null columns with outliers greater than 40% would be imputed using median

- Null columns with outliers less than 40% would be imputed using mean

```python
def z_prop(df, col_list):
    z_dict = {}
    for col in col_list:
#         origcol_size = df[col].shape[0]
        zs = df[(df[col] >=3) | (df[col]<=-3)]
        #display(zs)
        print(f'{col}  -   {zs.shape[0]}')
        prop = round((zs.shape[0]/df.shape[0])*100,2)
        z_dict[col] = prop
    return z_dict
```

```python
outlier_prop = z_prop(outliers, outliers.columns.to_list())
```

#### Proportion of outliers values in each column

```python
outlier_prop
```

```python
outlier_df = pd.Series(outlier_prop).reset_index()
outlier_df.rename(columns={'index':'features',0:'prop_outliers'}, inplace=True)
```

```python
outlier_df
```

```python
outlier_df.prop_outliers.unique()
```

##### None of our null columns have outlier values greater than 40% of normal values

- hence we would impute null values in this column using the mean, as we have some confidence that these outlier values would not greatly affect the mean
