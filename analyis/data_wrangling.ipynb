> We would detect outliers in each for the column by calculating thier Z-Score

```python
cols = list(numeric_data.columns)
outliers = pd.DataFrame()
#now iterate over the remaining columns and create a new zscore column
for col in cols:
    col_zscore = col + '_zscore'
    outliers[col_zscore] = (numeric_data[col] - numeric_data[col].mean())/numeric_data[col].std(ddof=0)
```

```python
outliers
```

#### values with __Z-score >=3 & Z-score <=-3__ are outliers.

 - If the frequency of such values are larger than 40% in any column, we would condier such column as having large number of outliers

- Null columns with outliers greater than 40% would be imputed using median

- Null columns with outliers less than 40% would be imputed using mean

```python
def z_prop(df, col_list):
    z_dict = {}
    for col in col_list:
#         origcol_size = df[col].shape[0]
        zs = df[(df[col] >=3) | (df[col]<=-3)]
        #display(zs)
        print(f'{col}  -   {zs.shape[0]}')
        prop = round((zs.shape[0]/df.shape[0])*100,2)
        z_dict[col] = prop
    return z_dict
```

```python
outlier_prop = z_prop(outliers, outliers.columns.to_list())
```
